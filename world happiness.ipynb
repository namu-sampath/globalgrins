```python
# add the new column using the mapping
happiness_df['region'] = happiness_df['country'].map(region_mapping)
happiness_df
```

```python
# what are the happiest countries in the world? (top 10; by happiness score)
happiest_countries = happiness_df.head(10)
happiest_countries
```

```python
# what countries are the unhappiest (by happiness score)
unhappiest_countries = happiness_df.tail(10)
unhappiest_countries
```

```python
col_names = happiness_df.columns

print("Names of all columns:")
print(col_names)
```

```python
# which countries have high gdp per capita

countries_with_high_gdp = []
for index, row in happiness_df.iterrows():
    if row['gdp per capita'] >= 1.0:
        countries_with_high_gdp.append(row['country'])
        
high_countries_string = ', '.join(countries_with_high_gdp)

print("Countries with GDP per capita greater than 1.0 and therefore economically stable are:", high_countries_string)
```

```python
len(countries_with_high_gdp)
```

```python
# which countries have low gdp per capita

countries_with_low_gdp = []
for index, row in happiness_df.iterrows():
    if row['gdp per capita'] < 1.0:
        countries_with_low_gdp.append(row['country'])
        
countries_string = ', '.join(countries_with_low_gdp)

print("The countries with GDP per capita less than 1.0 and therefore not economically stable are:", countries_string)
```

```python
len(countries_with_low_gdp)
```

```python
# which countries have a 0.000 gdp per capita score
economically_unstable = []
for index, row in happiness_df.iterrows():
    if row['gdp per capita'] == 0.0: 
        economically_unstable.append(row['country'])

countries_string = ', '.join(economically_unstable)

print(countries_string, "is the only country with a GDP per capita score of zero and therefore is economically unstable.")
```

```python
# which countries have a low gdp score and are unhappy 
low_gdp_countries = happiness_df[happiness_df['gdp per capita'] <= 1.0]
low_gdp_countries
```

```python
# filtering low GDP countries to get only the unhappiest ones (happiness score <= 4.0) 
unhappy_low_gdp_countries = low_gdp_countries[low_gdp_countries['happiness score'] <= 4.0]
unhappy_low_gdp_countries
```

```python
# getting the names of the countries that are both low GDP and unhappy
countries_low_gdp_unhappy = unhappy_low_gdp_countries['country'].tolist()
countries_low_gdp_unhappy

countries_string = ', '.join(countries_low_gdp_unhappy)

print("Countries with low GDP per capita and among the unhappiest are:", countries_string)
```

```python
# if a country has a perception of corruption score close to 0.0, it is 'very corrupt'
# how many countries are corrupt 

corrupt_countries = happiness_df[happiness_df['perceptions of corruption'] <= 0.10]
corrupt_countries
```

```python
corrupt_countries = []
for index, row in happiness_df.iterrows():
    if row['perceptions of corruption'] <= 0.10:
        corrupt_countries.append(row['country'])
        
corrupt_string = ', '.join(corrupt_countries)

print("These are the countries that are more likely to be corrupt:", corrupt_string)
```
